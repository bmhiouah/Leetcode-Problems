```python
def canPartitionSubset(nums,k):

    subsum = sum(nums)//k
    used = [False] * len(nums)

    def backtrack(i,k,subset):
        if k == 0:
            return True
        if subset == subsum:
            return backtrack(0,k-1,0)
        
        for j in range(i,len(nums)):
            if used[j] or subset + used[j] > subsum:
                continue
            used[j] = True
            if backtrack(j+1,k,subset+nums[j]):
                return True
            used[j] = False
        return False
    return backtrack(0,k,0)

exemple = [3,3,3,4,4,4]
canPartitionSubset(exemple,3)
```

```python
def reverse_Order_traversal(root):
    cumSum = 0

    def dfs(root):
        if not root:
            return
        nonlocal cumSum
        dfs(root.right)
        tmp = root.val
        root.val += cumSum
        cumSum += tmp
        dfs(root.left)
    dfs(root)
    return root

reverse_Order_traversal(tree).visualize_tree()
```
